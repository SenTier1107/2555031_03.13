```python
import seaborn as sns
import matplotlib.pyplot as plt

#
#seaborn tip 변수의 분석
#

# 데이터 로드
tips = sns.load_dataset('tips')

# 1. 기술 통계량 계산 [cite: 147, 238]
tip_mean = tips['tip'].mean()
tip_var = tips['tip'].var()
tip_skew = tips['tip'].skew() # 왜도 계산 [cite: 405]

print(f"Tip 평균: {tip_mean:.2f}")
print(f"Tip 분산: {tip_var:.2f}")
print(f"Tip 왜도(Skewness): {tip_skew:.2f}") # 0보다 크면 오른쪽으로 꼬리가 긴 분포

# 2. 분포 시각화 (히스토그램 + 밀도 함수)
plt.figure(figsize=(8, 5))
sns.histplot(tips['tip'], kde=True, color='skyblue')
plt.axvline(tip_mean, color='red', linestyle='--', label=f'Mean: {tip_mean:.2f}')
plt.axvline(tips['tip'].median(), color='green', linestyle='-', label=f'Median: {tips['tip'].median():.2f}')
plt.title('Distribution of Tip (Mean vs Median)')
plt.legend()
plt.show()

#분석: tip 데이터의 왜도(Skewness)가 약 1.47로 양수이므로, 오른쪽으로 꼬리가 긴 분포를 보임. 평균이 중앙값보다 높게 나타나는 이유는 고액의 팁(이상치)이 평균을 오른쪽으로 끌어당기기 때문임.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#
#평균의 함정 시각화
#

# 1. PDF 17p 예시 데이터 (마지막 값 248이 아주 작은 이상치 역할을 함)
data = [500, 489, 495, 493, 505, 248]
df_pitfall = pd.Series(data)

mean_val = df_pitfall.mean()   # 평균
median_val = df_pitfall.median() # 중앙값

print(f"평균(Mean): {mean_val:.2f}")
print(f"중앙값(Median): {median_val:.2f}")

# 2. 시각화로 비교
plt.figure(figsize=(10, 4))
plt.scatter(data, [1]*len(data), color='blue', label='Data Points')
plt.axvline(mean_val, color='red', linestyle='--', label=f'Mean ({mean_val:.2f})')
plt.axvline(median_val, color='green', linestyle='-', label=f'Median ({median_val:.2f})')
plt.title('The Pitfall of Mean: How Outliers Affect the Average')
plt.yticks([]) # y축 눈금 제거
plt.legend()
plt.show()

#분석 : 다른 데이터는 500 근처인데 하나만 248일 경우 평균은 455까지 떨어짐. 하지만 중앙값은 494로 데이터의 중심을 더 잘 유지함. 이것이 바로 "평균의 함정"이며, 이상치에 민감하지 않은 중앙값이 더 안전한 지표가 될 수 있음.
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. 데이터 로드 (모집단)
tips = sns.load_dataset('tips')
pop_mean = tips['tip'].mean()
pop_var = tips['tip'].var()

# 2. 3가지 표본추출 (샘플 사이즈는 약 50개 정도로 설정)
# (1) 단순무작위추출 (SRS)
srs_sample = tips.sample(n=50, random_state=42)

# (2) 층화무작위추출 (요일별로 비율에 맞게 추출)
stratified_sample = tips.groupby('day', group_keys=False).apply(lambda x: x.sample(frac=0.2, random_state=42))

# (3) 집락추출 (요일 중 하나를 무작위로 선택하여 해당 요일 전체 조사)
days = tips['day'].unique()
selected_day = np.random.choice(days)
cluster_sample = tips[tips['day'] == selected_day]

# 3. 평균 및 분산 비교 데이터프레임 만들기
results = pd.DataFrame({
    'Method': ['Population', 'SRS', 'Stratified', 'Cluster'],
    'Mean': [pop_mean, srs_sample['tip'].mean(), stratified_sample['tip'].mean(), cluster_sample['tip'].mean()],
    'Variance': [pop_var, srs_sample['tip'].var(), stratified_sample['tip'].var(), cluster_sample['tip'].var()]
})

# 4. 시각화 (그래프로 비교)
plt.figure(figsize=(12, 5))

# 평균 비교 그래프
plt.subplot(1, 2, 1)
sns.barplot(x='Method', y='Mean', data=results, palette='viridis')
plt.axhline(pop_mean, color='red', linestyle='--', label='Pop Mean')
plt.title('Mean Comparison')

# 분산 비교 그래프
plt.subplot(1, 2, 2)
sns.barplot(x='Method', y='Variance', data=results, palette='magma')
plt.axhline(pop_var, color='red', linestyle='--', label='Pop Var')
plt.title('Variance Comparison')

plt.tight_layout()
plt.show()

print(results)
```
